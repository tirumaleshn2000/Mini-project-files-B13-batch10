```python
import urllib.request as urlrq
import certifi
#loading the data from url.
resp = urlrq.urlopen('https://data.covid19india.org/csv/latest/state_wise_daily.csv', cafile=certifi.where())
```

```python
#reading the data as dataframe.
data_frame=pd.read_csv(resp)
```

```python
data_frame.head()
```

```python
recovered_data=data_frame[data_frame['Status']=='Recovered']
confirmed_data=data_frame[data_frame['Status']=='Confirmed']
deceased_data=data_frame[data_frame['Status']=='Deceased']
```

```python
recovered_data.to_csv('recovered_data.csv',index=False)
confirmed_data.to_csv('daily cases forecasting/confirmed_data.csv',index=False)
deceased_data.to_csv('deceased_data.csv',index=False)
```
